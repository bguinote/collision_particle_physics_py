```python
import numpy as np
import matplotlib.pylab as plt
from math import sqrt
import pps_tools as pps
import h5hep
```

```python
pps.download_from_drive('dimuons_1000_collisions.hdf5')
```

```python
# biblioteca pega o arquivo e baixa o arquivo e pega os dados.
```

```python
infile = 'data/dimuons_1000_collisions.hdf5'
collisions = pps.get_collisions(infile,experiment='CMS',verbose=False)
```

```python
# calculando quadrimomentum e magnitude do momentum
```

```python
e = []
px = []
py = []
pz = []

for collision in collisions:
    x = collision["muons"]
    for muon in x: #pegando dados de todos os muons e anti-muons
        e.append (muon["e"])
        px.append (muon["px"])
        py.append (muon["py"])
        pz.append (muon["pz"])
```

```python
print(e)
```

```python
p = []

for index in range(0, len(px)):
    momentum = sqrt(px[index] ** 2 + py[index] ** 2 + pz[index] ** 2) # magnitude do momento
    p.append(momentum)
```

```python
print(p)
```

```python
plt.hist(p,bins=100,range=(0,100))
plt.title("Colisão Próton-Próton")
plt.xlabel('Momentum')
plt.ylabel('Partículas')
```

```python
plt.hist(p,bins=100,range=(0,100))
plt.hist(e,bins=100,range=(0,100))
plt.title("Colisão Próton-Próton")
plt.xlabel('Energia')
plt.ylabel('Partículas')
```

```python
m = []

for index in range(0, len(px)):
    massa = p[index]**2/2*e[index] # massa do muon
    m.append(massa)
```
